```python
# 2628
N, M = map(int, input().split())
row = [0, N]
col = [0, M]
k = int(input())
for _ in range(k):
    a, b = map(int, input().split())
    #col
    if a == 0:
        col.append(b)
    #row
    else:
        row.append(b)

row.sort()
col.sort()
cand = []
for i in range(len(row)-1):
    for j in range(len(col)-1):
        cand.append((row[i+1]-row[i])*(col[j+1]-col[j]))

# print(cand)
print(max(cand))

```
